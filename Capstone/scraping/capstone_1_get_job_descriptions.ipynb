```python
from bs4 import BeautifulSoup as Soup
import urllib, requests, re, pandas as pd
import numpy as np
import time

#load jobs metadata with links that we will access for the descriptions
df2 = pd.read_csv("job_ad_metadata_v2.csv")
df2["description"] = None
# Trying to include full page in dataframe misses with saving to .csv, omit it for now
#df2["job_page"] = None
# create column that flags jobs that couldn't be accessed due to errors
df2["error"] = 0
```

```python
# Iterate over jobs and get job page and extract description
for i in df2.index:
    if (i % 10) == 0:
        print(i)
    time.sleep(np.random.randint(low = 1, high = 4))
    try:
        # Access page
        target_job = Soup(urllib.request.urlopen(df2["job_link"][i]), "lxml")
        
        # remove extraneous elements
        for elem in target_job.find_all(["script", "style"]):
            elem.extract()
        
        # Put description and page in dataframe
        df2.loc[i, "description"] = target_job.get_text()
        #df2.loc[i, "job_page"] = str(target_job)
        
    except (urllib.error.HTTPError, urllib.error.URLError, ConnectionResetError):
        # Record in dataframe if page can't be accessed
        print(i)
        print("error")
        df2.loc[i, "description"] = None
        df2.loc[i, "error"] = 1
        

df2.to_csv("job_ad_descriptions_v2.csv", index = False)
df2.head()
```
